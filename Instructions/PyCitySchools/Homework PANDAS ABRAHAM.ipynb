```python
#update names of columns in school data

schoolsdataorg_df = schoolsdata_file_pd.rename(columns={"name":"School Name", "type":"School Type",
                                                       "size":"School Size", "budget":"School Budget"})
schoolsdataorg_df.head()
```

```python
#update formatting of School Size and School Budget

schoolsdataorg_df["School Size"] = schoolsdataorg_df["School Size"].map("{:,}".format)


schoolsdataorg_df.head()
```

```python
#read students_data csv file

studentsdata_file_pd = pd.read_csv(students_data)
studentsdata_file_pd.head()
```

```python
studentsdataorg_df = studentsdata_file_pd.rename(columns={"name":"Student Name", "gender":"Gender",
                                                       "grade":"Grade Level", "school":"School Name",
                                                       "reading_score":"Reading Score", "math_score":"Math Score"})

studentsdataorg_df.head()
```

```python
student_school_merge_df = pd.merge(studentsdataorg_df, schoolsdataorg_df, how="left", on=["School Name", "School Name"])

student_school_merge_df.head(25)
```

```python
# District Summary
# define variables

total_students = len(studentsdataorg_df)

total_budget = schoolsdataorg_df["School Budget"].sum()

mathscore_avg = round(studentsdataorg_df["Math Score"].mean(), 2)
readscore_avg = round(studentsdataorg_df["Reading Score"].mean(), 2)

mathpass = len(studentsdataorg_df.loc[studentsdataorg_df['Math Score'] >= 70])
percent_mathpass = round(mathpass/total_students *100, 2)

readpass = len(studentsdataorg_df.loc[studentsdataorg_df['Reading Score'] >= 70])
percent_readpass = round(readpass/total_students *100, 2)

district_passrate = round((percent_mathpass + percent_readpass)/2, 2)
```

```python
# Placed all of the data found into a summary DataFrame for "District Summary"
district_summary_table = ({"Total Students": [total_students],
                              "Total Budget": [total_budget],
                              "Average Math Score": [mathscore_avg],
                              "Average Reading Score": [readscore_avg],
                                "Percent Passing Math": [percent_mathpass],
                                "Percent Passing Reading": [percent_readpass],
                                "Overall Passing Rate": [district_passrate]
                                })
print("District Summary")
district_summary_table
```

```python
df_districtsummary = pd.DataFrame([district_summary_table])

df_districtsummary
```

```python
# Create list of school, type

eachschool_namelisttype = schoolsdataorg_df .set_index(["School Name"])["School Type"]

eachschool_namelisttype
```

```python
# total students
eachschool_totalstudents = student_school_merge_df["School Name"].value_counts()
eachschool_totalstudents
```

```python
# total budget
eachschool_namelist_budget = schoolsdataorg_df.groupby(["School Name"]).sum()["School Budget"]

eachschool_namelist_budget
```

```python
# budget per student

eachschool_budgetperstudent = eachschool_namelist_budget / eachschool_totalstudents

eachschool_budgetperstudent
```

```python
# average math scores by school
eachschool_mathaverage = student_school_merge_df.groupby(["School Name"]).mean()["Math Score"]

eachschool_mathaverage
```

```python
# average reading scores by school

eachschool_readaverage = student_school_merge_df.groupby(["School Name"]).mean()["Reading Score"]

eachschool_readaverage
```

```python
# Percent Passing Math by School

school_mathscores_list70 = student_school_merge_df[(student_school_merge_df["Math Score"] >= 70)]
eachschool_percent_mathpass = school_mathscores_list70.groupby(["School Name"]).count()["Student Name"] / eachschool_totalstudents *100

eachschool_percent_mathpass

```

```python
# Percent Passing Reading by School

school_readscores_list70 = student_school_merge_df[(student_school_merge_df["Reading Score"] >= 70)]
eachschool_percent_readpass = school_readscores_list70.groupby(["School Name"]).count()["Student Name"] / eachschool_totalstudents *100

eachschool_percent_readpass
```

```python
# Overall Passing Rate by School

eachschool_passrate = round((eachschool_percent_mathpass+eachschool_percent_readpass)/2, 2)



eachschool_passrate
```

```python
eachschool_summary_table = ({"School Type": eachschool_namelisttype,
                                "School Total Students":eachschool_totalstudents,
                                "School Total Budget": eachschool_namelist_budget,
                                "School Per Student Budget": eachschool_budgetperstudent,
                                "School Average Math Score": eachschool_mathaverage,
                                "School Average Reading Score":eachschool_readaverage,
                                "School Percent Passing Math": eachschool_percent_mathpass,
                                "School Percent Passing Reading": eachschool_percent_readpass,
                                "School Overall Passing Rate": eachschool_passrate})

df_eachschool_summary = pd.DataFrame(eachschool_summary_table)

df_eachschool_summary.head()


df_eachschool_summary = df_eachschool_summary[["School Type", "School Total Students","School Total Budget", "School Per Student Budget",
                                               "School Average Math Score", "School Average Reading Score", "School Percent Passing Math",
                                               "School Percent Passing Reading", "School Overall Passing Rate"]]

df_eachschool_summary 
```

```python
df_eachschool_summary.dtypes
```

```python
# Top 5 Performing Schools (By Passing Rate)**
df_top5_schools = df_eachschool_summary.sort_values("School Overall Passing Rate", ascending=False)
df_top5_schools.head()
```

```python
# **Bottom 5 Performing Schools (By Passing Rate)**
df_bottom5_schools = df_eachschool_summary.sort_values("School Overall Passing Rate", ascending=True)
df_bottom5_schools.head()
```

```python
# **Math Scores by Grade**
df_students_bygrade = student_school_merge_df.groupby("Grade Level")
df_avgmath_bygrade = df_students_bygrade["Math Score"].mean()
df_avgmath_bygrade
```

```python
# **Reading Scores by Grade**
df_students_bygrade = student_school_merge_df.groupby("Grade Level")
df_avgread_bygrade = df_students_bygrade["Reading Score"].mean()
df_avgread_bygrade
```

```python
# **Scores by School Spending**

# * Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
#   * Average Math Score
#   * Average Reading Score
#   * % Passing Math
#   * % Passing Reading
#   * Overall Passing Rate (Average of the above two)

schoolspend_bins = [550, 580, 610, 640, 700]
schoolspend_tierlabels = ["Tier 1 - LOW", "Tier 2 - Low Mid", "Tier 3 - High Mid", "Tier 4 - HIGH"]

df_eachschool_summary["Spend Tier"] = pd.cut(df_eachschool_summary["School Per Student Budget"], schoolspend_bins, labels=schoolspend_tierlabels)
schoolspend_df = df_eachschool_summary[["Spend Tier", "School Average Math Score", "School Average Reading Score", 
                                             "School Percent Passing Math", "School Percent Passing Reading", "School Overall Passing Rate"]].copy()
schoolspend_sort_df = schoolspend_df.sort_values("Spend Tier", ascending=False)
schoolspend_sort_df
```

```python
# **Scores by School Size**

# * Repeat the above breakdown, but this time group schools based on a reasonable approximation of 
# school size (Small, Medium, Large).

schoolsize_bins = [0, 2000, 4000, 6000]
schoolsize_tierlabels = ["Small", "Medium", "Large"]

df_eachschool_summary["Size of School"] = pd.cut(df_eachschool_summary["School Total Students"], schoolsize_bins, labels=schoolsize_tierlabels)
schoolsize_df = df_eachschool_summary[["Size of School", "School Average Math Score", "School Average Reading Score", 
                                             "School Percent Passing Math", "School Percent Passing Reading", "School Overall Passing Rate"]].copy()
schoolsize_sort_df = schoolsize_df.sort_values("Size of School", ascending=False)
schoolsize_sort_df
```

```python
# **Scores by School Type**

# * Repeat the above breakdown, but this time group schools based on school type (Charter vs. District).

schooltype_df = df_eachschool_summary[["School Type", "School Average Math Score", "School Average Reading Score", 
                                             "School Percent Passing Math", "School Percent Passing Reading", 
                        "School Overall Passing Rate"]].copy() 
schooltype_sort_df = schooltype_df.sort_values("School Type", ascending=False)
schooltype_sort_df
```
